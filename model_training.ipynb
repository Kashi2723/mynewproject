```python
num_coltypes
```

```python
newadult=[False, True]
```

```python
catpipe = Pipeline(
    steps=[
        ('encoder',OrdinalEncoder(categories=[newsex,newclass,newadult])),
        ('scaler',StandardScaler())
    ]
)
```

```python
numpipe = Pipeline(
    steps=[
        ('imputer',SimpleImputer()),
        ('scaler',StandardScaler())
    ]
)
```

```python
cat_coltypes
```

```python
preprocessor = ColumnTransformer(

    transformers= [
        ('cat_trans',catpipe,cat_col.columns),
        ('num_trans',numpipe,num_col.columns)
    ]
)
```

```python
xtrain = pd.DataFrame(preprocessor.fit_transform(xtrain), columns= preprocessor.get_feature_names_out())
xtrain
```

```python
xtest = pd.DataFrame(preprocessor.transform(xtest), columns=preprocessor.get_feature_names_out())
xtest
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, root_mean_squared_error
```

```python
models = {
    'linear': LinearRegression(),
    'Ridge' : Ridge(),
    'Lasso' : Lasso(),
    'ElasticNet' : ElasticNet()
}
```

```python
for model in models.keys():
    models[model].fit(xtrain,ytrain)
    ypred = models[model].predict(xtest)

    print(model)
    print('mean_squared_error',mean_squared_error(ytest,ypred))
    print('mean_absolute_error',mean_absolute_error(ytest,ypred))
    print('r2_score',r2_score(ytest,ypred))
    print('root_mean_squared_error',root_mean_squared_error(ytest,ypred))

    print("="*100)
```
